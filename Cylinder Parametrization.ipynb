```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
ax=Axes3D(fig,azim=0,elev=0)

# Intersection curve!!!
u = np.linspace(-np.pi,np.pi,100)
v = np.linspace(-np.pi,np.pi,100)
#first, make a circle with form x = r*sin(u), y = r*cos(u)
x = 1.5*np.sin(u)
y = 1.5*np.cos(u)
#then PROJECT onto cylinder z = (sqrt(a-(x-d)^2) - by + c. a is the radius, b is an angular offset, and
#c is a translation (which is IMPORTANT for something).
z = np.sqrt(1-(x-0)**2)-0*y+0


# Cylinder
u1,v1 = np.mgrid[0:2*np.pi:30j,-np.pi:np.pi:30j]
x1 = 1.5*np.sin(u1)
y1 = 1.5*np.cos(u1)
z1 = v1

#rotated cylinder
x2 = np.cos(u1)
y2 = -.707*np.sin(u1) - .707*v1
z2 = -.707*np.sin(u1) + .707*v1


ax.plot_surface(x1, y1, z1,  rstride=4, cstride=4, color = 'blue', alpha = 0.25, lw = 0)
ax.plot_surface(x2, y2, z2,  rstride=4, cstride=4, color = 'green', alpha = 0.25, lw = 0)
ax.plot_wireframe(x, y, z,  rstride=4, cstride=4, cmap = cm.copper, color = 'red', lw = 5, linestyle = '--')

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.show()
```
